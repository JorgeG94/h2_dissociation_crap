```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("results.csv")

# Extract bond distances
x = df["R"]

# Start the plot
plt.figure(figsize=(10, 6))
plt.title("H₂ Potential Energy Curves", fontsize=16)
plt.xlabel("Bond Distance R (Å)", fontsize=14)
plt.ylabel("Energy (Hartree)", fontsize=14)

# Loop through each method/functional and plot if data is present
for col in df.columns[1:]:
    y = pd.to_numeric(df[col], errors='coerce')  # Convert invalids to NaN
    plt.plot(x, y, label=col)

# Add legend
plt.legend(title="Method / Functional", fontsize=10, loc='best')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
